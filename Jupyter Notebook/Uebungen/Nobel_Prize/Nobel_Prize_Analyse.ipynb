```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
import numpy as np
import pandas as pd

sns.set()

archive = pd.read_csv("archive.csv", sep = ',')
archive['Year'].value_counts()
_ = sns.swarmplot(x='Year', data=archive)
# Label the axes
_ = plt.xlabel('art of iris')
_ = plt.ylabel('petal length (cm)')
plt.show()
```
